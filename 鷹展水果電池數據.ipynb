```python
def MakePlot(t, y, y_max, name):
    x = t
    colors = ['r', 'g', 'b']  # 設定每個資料集的顏色
    labels = ['X軸', 'Y軸', 'Z軸']  # 每個資料集的標籤
    
    plt.figure(figsize=(10, 5))
    plt.xticks(np.arange(0, 6, 0.5))  # 設定 X 軸刻度
    
    for i in range(len(y)):
        # 擬合每個資料集
        m, b = np.polyfit(x, y[i], 1)
        y_fit = m * np.linspace(min(x), max(x), 100) + b  # 產生平滑的趨勢線
        
        # 計算 MSE
        y_pred = m * x + b
        mse = np.mean((y[i] - y_pred) ** 2)
        
        # 繪製散點圖與趨勢線
        plt.scatter(x, y[i], marker="+", color=colors[i])  # 散點圖
        plt.plot(np.linspace(min(x), max(x), 100), y_fit, color = colors[i], label=f"{name}{labels[i]}趨勢 : y = {m:.2f}x + {b:.2f}")  # 趨勢線與 MSE
    
    # 圖表設置
    plt.xlabel("時間(min)")  # X 軸標題
    plt.ylabel("電壓(V)")  # Y 軸標題
    plt.title(f"{name}電壓-時間")  # 圖表標題
    plt.legend()  # 顯示圖例
    plt.xlim([-0.2, 5.2])  # 固定 X 軸上下限
    plt.ylim([0.3, y_max])  # 固定 Y 軸上下限
    plt.show()

t = np.array(time)
y = [np.array(Total[2]),
     np.array(Total[3]),
     np.array(Total[4])]
y_max = 0.7
name = "奇異果"

# 呼叫函式
MakePlot(t, y, y_max, name)

```
